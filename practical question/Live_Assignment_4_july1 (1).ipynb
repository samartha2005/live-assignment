```python
'''1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*) .'''
#Code:
def calculator(operation, num1, num2):
    if operation == '+':
        result = num1 + num2
    elif operation == '-':
        result = num1 - num2
    elif operation == '*':
        result = num1 * num2
    elif operation == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            result = "Error: Division by zero"
    else:
        result = "Error: Invalid operation"
    
    return result

# Example usage:
print(calculator('+', 5, 3))   # Output: 8 (5 + 3)
print(calculator('-', 7, 2))   # Output: 5 (7 - 2)
print(calculator('*', 4, 6))   # Output: 24 (4 * 6)
print(calculator('/', 10, 2))  # Output: 5.0 (10 / 2)
print(calculator('/', 5, 0))   # Output: Error: Division by zero
print(calculator('%', 3, 4))   # Output: Error: Invalid operation
```

```python
'''1.9. Write a program to find the factors of a whole number using a while loop.'''
#Code:
def find_factors(number):
    i = 1
    factors = []
    while i <= number:
        if number % i == 0:
            factors.append(i)
        i += 1
    return factors

# Input from the user
num = int(input("Enter a whole number: "))

# Finding factors
factors = find_factors(num)

# Displaying the factors
print(f"The factors of {num} are: {factors}")
```

```python
'''1.11. Write a program to find prime numbers between 2 to 100 using nested for loops.'''
#Code:
# Function to find and print prime numbers between 2 and 100
def find_primes():
    for num in range(2, 101):  # Loop through numbers from 2 to 100
        is_prime = True
        for i in range(2, int(num**0.5) + 1):  # Check divisibility up to the square root of the number
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=' ')

# Calling the function to print prime numbers between 2 and 100
find_primes()
```

```python
'''1.12. Write the programs for the following
 Accept the marks of the student in five major subjects and display the same
 Calculate the sum of the marks of all subjects.Divide the total marks by number of subjects (i.e. 5), calculate
 percentage = total marks and display the percentage.
 Find the grade of the student as per the following criteria . Hint: Use Match & case for this.:
'''

#Code:
def get_grade(percentage):
    match percentage:
        case p if p > 85:
            return 'A'
        case p if 75 <= p <= 85:
            return 'B'
        case p if 50 <= p < 75:
            return 'C'
        case p if 30 <= p < 50:
            return 'D'
        case p if p < 30:
            return 'Reappear'

# Function to accept marks and calculate grade
def student_marks_and_grade():
    subjects = ['Subject 1', 'Subject 2', 'Subject 3', 'Subject 4', 'Subject 5']
    marks = []
    
    # Accept marks for each subject
    for subject in subjects:
        mark = float(input(f"Enter marks for {subject}: "))
        marks.append(mark)
    
    # Display the marks
    print("\nMarks obtained:")
    for subject, mark in zip(subjects, marks):
        print(f"{subject}: {mark}")
    
    # Calculate total marks and percentage
    total_marks = sum(marks)
    percentage = total_marks / len(subjects)
    
    # Display total marks and percentage
    print(f"\nTotal Marks: {total_marks}")
    print(f"Percentage: {percentage}%")
    
    # Determine and display grade
    grade = get_grade(percentage)
    print(f"Grade: {grade}")

# Run the function
student_marks_and_grade()

```

```python
'''4.Create a simple baning system using object-oriented concepts in Python. Design classes representing
different types of ban accounts such as savings and checing. Implement methods for deposit, withdraw,
and balance inquiry. Utilize inheritance to manage different account types efficiently.
Tasks
 1. Define base class(es) for bank accounts with common attributes and methods
 2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount)
 3. Provide methods for deposit, withdraw, and balance inquiry in each subclass
 4. Test the banking system by creating instances of different account types and performing transactions.'''
# code:
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount:.2f}. New balance is ${self.__balance:.2f}.")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount:.2f}. New balance is ${self.__balance:.2f}.")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    # Method to check balance
    def check_balance(self):
        return self.__balance

    # Method to display account information
    def display_info(self):
        print(f"Account Number: {self.__account_number}")
        print(f"Account Holder: {self.__account_holder}")
        print(f"Balance: ${self.__balance:.2f}")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.02):
        super().__init__(account_number, account_holder, balance)
        self.__interest_rate = interest_rate

    # Method to apply interest
    def apply_interest(self):
        interest = self.check_balance() * self.__interest_rate
        self.deposit(interest)
        print(f"Interest of ${interest:.2f} applied at rate {self.__interest_rate*100:.2f}%.")

    # Overriding display_info method
    def display_info(self):
        super().display_info()
        print(f"Interest Rate: {self.__interest_rate*100:.2f}%")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, overdraft_limit=500.0):
        super().__init__(account_number, account_holder, balance)
        self.__overdraft_limit = overdraft_limit

    # Overriding withdraw method to handle overdraft
    def withdraw(self, amount):
        if amount > 0 and amount <= (self.check_balance() + self.__overdraft_limit):
            super().withdraw(amount)
        else:
            print(f"Invalid withdrawal amount. Overdraft limit of ${self.__overdraft_limit:.2f} exceeded.")

    # Overriding display_info method
    def display_info(self):
        super().display_info()
        print(f"Overdraft Limit: ${self.__overdraft_limit:.2f}")
# Creating instances and testing the functionality

# Creating a savings account
savings = SavingsAccount("SA123", "Alice", 1000.0, 0.03)
print("Savings Account Information:")
savings.display_info()
print("\nDepositing $500 to Savings Account:")
savings.deposit(500)
print("\nApplying Interest to Savings Account:")
savings.apply_interest()
print("\nWithdrawing $200 from Savings Account:")
savings.withdraw(200)
print("\nFinal Savings Account Information:")
savings.display_info()

# Creating a checking account
checking = CheckingAccount("CA123", "Bob", 500.0, 1000.0)
print("\n\nChecking Account Information:")
checking.display_info()
print("\nDepositing $300 to Checking Account:")
checking.deposit(300)
print("\nWithdrawing $700 from Checking Account:")
checking.withdraw(700)
print("\nAttempting to Withdraw $1200 from Checking Account:")
checking.withdraw(1200)
print("\nFinal Checking Account Information:")
checking.display_info()
```

```python
'''7.Write a code for Hotel Management System using OOPS
1 Create a Room class that has attributes such as room number, room type, rate, and availability (private)
2 Implement methods to book a room, check in a guest, and check out a guest
3 Use encapsulation to hide the room's unique identifi'ation number
4 Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, ea'h with their own
specific attributes and methods.'''
# Code:
class Room:
    _id_counter = 1  # Class-level attribute to generate unique IDs

    def __init__(self, room_number, room_type, rate, available=True):
        self.__id = Room._id_counter
        Room._id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__available = available

    # Getter for ID
    def get_id(self):
        return self.__id

    # Method to book a room
    def book_room(self):
        if self.__available:
            self.__available = False
            print(f"Room {self.room_number} has been booked.")
        else:
            print(f"Room {self.room_number} is already booked.")

    # Method to check in a guest
    def check_in(self):
        if not self.__available:
            print(f"Guest checked into Room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    # Method to check out a guest
    def check_out(self):
        if not self.__available:
            self.__available = True
            print(f"Guest checked out of Room {self.room_number}. Room is now available.")
        else:
            print(f"Room {self.room_number} is already available.")

    # Method to display room information
    def display_info(self):
        availability = "Available" if self.__available else "Booked"
        print(f"ID: {self.__id}")
        print(f"Room Number: {self.room_number}")
        print(f"Room Type: {self.room_type}")
        print(f"Rate: ${self.rate:.2f}")
        print(f"Availability: {availability}")

# Subclass: SuiteRoom
class SuiteRoom(Room):
    def __init__(self, room_number, rate, available=True, has_living_area=True):
        super().__init__(room_number, "Suite", rate, available)
        self.has_living_area = has_living_area

    # Overriding display_info method to include specific attribute
    def display_info(self):
        super().display_info()
        print(f"Has Living Area: {'Yes' if self.has_living_area else 'No'}")

# Subclass: StandardRoom
class StandardRoom(Room):
    def __init__(self, room_number, rate, available=True, bed_type="Queen"):
        super().__init__(room_number, "Standard", rate, available)
        self.bed_type = bed_type

    # Overriding display_info method to include specific attribute
    def display_info(self):
        super().display_info()
        print(f"Bed Type: {self.bed_type}")

# Hotel Management System
class Hotel:
    def __init__(self):
        self.rooms = []

    # Method to add a new room
    def add_room(self, room):
        self.rooms.append(room)
        print(f"Added Room {room.room_number} to the hotel.")

    # Method to book a room
    def book_room(self, room_id):
        for room in self.rooms:
            if room.get_id() == room_id:
                room.book_room()
                return
        print(f"Room ID {room_id} not found.")

    # Method to check in a guest
    def check_in_guest(self, room_id):
        for room in self.rooms:
            if room.get_id() == room_id:
                room.check_in()
                return
        print(f"Room ID {room_id} not found.")

    # Method to check out a guest
    def check_out_guest(self, room_id):
        for room in self.rooms:
            if room.get_id() == room_id:
                room.check_out()
                return
        print(f"Room ID {room_id} not found.")

    # Method to display all rooms
    def display_rooms(self):
        print("\nHotel Rooms:")
        for room in self.rooms:
            room.display_info()
            print()

# Testing the Hotel Management System
hotel = Hotel()

# Adding rooms to the hotel
suite1 = SuiteRoom(101, 200.0)
suite2 = SuiteRoom(102, 250.0, has_living_area=False)
standard1 = StandardRoom(201, 100.0)
standard2 = StandardRoom(202, 120.0, bed_type="King")

hotel.add_room(suite1)
hotel.add_room(suite2)
hotel.add_room(standard1)
hotel.add_room(standard2)

# Displaying all rooms
hotel.display_rooms()

# Booking a room
hotel.book_room(suite1.get_id())

# Checking in a guest
hotel.check_in_guest(suite1.get_id())

# Checking out a guest
hotel.check_out_guest(suite1.get_id())

# Displaying all rooms after transactions
hotel.display_rooms()
```

```python
'''19. Create a Python program that reads a text file named "paragraph.txt" and counts the occurrences of
each word in the paragraph, displaying the results in alphabetical order.'''
#Code
import string
from collections import defaultdict

def count_word_occurrences(file_path):
    word_count = defaultdict(int)

    try:
        with open(file_path, 'r') as file:
            # Read the entire file content
            content = file.read()

            # Remove punctuation and convert to lowercase
            translator = str.maketrans('', '', string.punctuation)
            content = content.translate(translator).lower()

            # Split the content into words
            words = content.split()

            # Count occurrences of each word
            for word in words:
                word_count[word] += 1

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

    return word_count

def main():
    file_path = "paragraph.txt"
    word_count = count_word_occurrences(file_path)

    if word_count is not None:
        # Display word counts in alphabetical order
        for word in sorted(word_count.keys()):
            print(f"{word}: {word_count[word]}")

if __name__ == "__main__":
    main()
```

```python
'''21 What do you mean by skewness.Explain its types.Use graph to show.'''
#Ans:
'''Skewness in statistics refers to the measure of asymmetry or lack of symmetry in the distribution of data. 
It indicates whether the data is concentrated more on one side than the other relative to the mean. 
A perfectly symmetrical distribution has zero skewness.

Types of Skewness:

Positive Skewness (Right Skewness):
In a positively skewed distribution, the tail of the distribution extends towards the right side. 
This means that there are more extreme values on the right side of the distribution (higher values) compared to the left.
The mean is typically greater than the median and mode in a positively skewed distribution.
Example: Income distribution, where a few individuals earn significantly higher incomes than the majority.

Negative Skewness (Left Skewness):
In a negatively skewed distribution, the tail of the distribution extends towards the left side. 
This indicates that there are more extreme values on the left side of the distribution (lower values) compared to the right.
The mean is typically less than the median and mode in a negatively skewed distribution.
Example: Test scores of a difficult exam, where most students perform well but a few perform poorly.

Explanation:
Positive Skewness Graph: The histogram on the left shows a positively skewed distribution. Note the longer tail on the
right side (higher values).

Negative Skewness Graph: The histogram on the right shows a negatively skewed distribution. Here, the tail extends more 
towards the left side (lower values).

Interpretation:
Skewness Measure: Skewness can be quantitatively measured using statistical formulas, such as Pearson's 
first skewness coefficient or the skewness function in Python libraries like scipy.stats.

Implications: Understanding skewness helps in interpreting data distributions accurately. 
It informs decisions in fields like finance, economics, and social sciences, 
where understanding the shape and direction of data distribution is crucial for analysis and decision-making.'''

#Code
import numpy as np
import matplotlib.pyplot as plt

# Generate data for positive skewness
np.random.seed(0)
data_pos_skew = np.random.exponential(scale=1.0, size=1000)  # Exponential distribution

# Generate data for negative skewness
data_neg_skew = 10 - np.random.exponential(scale=1.0, size=1000)  # Mirrored exponential distribution

# Plotting histograms
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(data_pos_skew, bins=30, color='skyblue', edgecolor='black')
plt.title('Positive Skewness')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data_neg_skew, bins=30, color='salmon', edgecolor='black')
plt.title('Negative Skewness')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
