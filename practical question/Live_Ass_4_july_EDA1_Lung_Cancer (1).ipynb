```python
# importing some libraryies for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Class distribution
class_distribution = df['LUNG_CANCER'].value_counts()
print("\nClass Distribution:")
print(class_distribution)
```

```python
# Correlation matrix
correlation_matrix = df.corr()
correlation_matrix
```

```python
# Plotting the correlation matrix
plt.figure(figsize=(12, 10),facecolor='lightgray')

# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True, linewidths=.5)
plt.title('Correlation Matrix of Numerical Features', color='#F81313',  fontproperties=font, fontsize= 18)
plt.show()
```

```python
# Select numerical features
num_features = df.select_dtypes(include=['int64', 'float64']).columns

# Set the font properties
font = {'family': 'serif', 'style': 'italic', 'weight': 'bold'}

# Create a figure with subplots
plt.figure(figsize=(20, 20), facecolor='lightgray')
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Determine the number of rows and columns for the subplots
num_rows = len(num_features) // 3 + (len(num_features) % 3 > 0)
num_cols = 3

# Loop through the numerical features and create a histogram for each
for idx, feature in enumerate(num_features):
    plt.subplot(num_rows, num_cols, idx + 1)
    sns.histplot(df[feature], kde=True, bins=20, color='#6AE89B')
    plt.title(f'Distribution of {feature}', color='#F81313', fontdict=font, fontsize=14)

# Show the plot
plt.show()
```

```python
# Gender distribution
gender_distribution = df['GENDER'].value_counts()
gender_distribution
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution in Lung Cancer Dataset')
plt.show()
```

```python
# Select numerical features
num_features = ['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
                'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
                'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']

# Melt the dataset for seaborn's boxplot
melted_data = df.melt(id_vars=['LUNG_CANCER'], value_vars=num_features, var_name='Feature', value_name='Value')

# Set the font properties
font = {'family': 'serif', 'style': 'italic', 'weight': 'bold'}

# Create the box plot
plt.figure(figsize=(20, 10))
sns.boxplot(x='Feature', y='Value', hue='LUNG_CANCER', data=melted_data)
plt.xticks(rotation=90)
plt.title('Features vs LUNG_CANCER', fontdict=font)
plt.show()
```

```python
# Additional insights
# Example: Analyzing the relationship between age and lung cancer
plt.figure(figsize=(10, 5),facecolor='lightgrey')
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here
plt.xlabel('Age', color='#5CD3E6',  fontproperties=font, fontsize= 16)
plt.ylabel('Count', color='#5CD3E6',  fontproperties=font, fontsize= 16)
sns.histplot(df[df['LUNG_CANCER'] == 'YES']['AGE'], color='red', kde=True, label='LUNG_CANCER: YES')
sns.histplot(df[df['LUNG_CANCER'] == 'NO']['AGE'], color='blue', kde=True, label='LUNG_CANCER: NO')
plt.legend()
plt.title('Age Distribution by Lung Cancer Status',color='#3CEE1C',  fontproperties=font, fontsize= 18)
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind
num_features = df.select_dtypes(include=['int64', 'float64']).columns

# Separate data into two groups based on lung cancer diagnosis
group_yes = df[df['LUNG_CANCER'] == 'YES']
group_no = df[df['LUNG_CANCER'] == 'NO']

# Perform t-test for each numerical feature
t_test_results = {}
for feature in num_features:
    t_stat, p_value = ttest_ind(group_yes[feature], group_no[feature], equal_var=False)
    t_test_results[feature] = {'t_stat': t_stat, 'p_value': p_value}

# Convert results to a DataFrame for better visualization
t_test_df = pd.DataFrame(t_test_results).T
t_test_df.columns = ['t_statistic', 'p_value']

# Display the results
print(t_test_df)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

num_features = df.select_dtypes(include=['int64', 'float64']).columns

# Perform ANOVA for each numerical feature
anova_results = {}
for feature in num_features:
    formula = f'{feature} ~ C(LUNG_CANCER)'
    model = ols(formula, data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    anova_results[feature] = anova_table

# Display the results
for feature, result in anova_results.items():
    print(f'\nANOVA results for {feature}:')
    print(result)
```
