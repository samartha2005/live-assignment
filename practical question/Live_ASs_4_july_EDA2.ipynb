```python
# importing some libraryies for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("election2024.csv")
# Display the first few rows of the dataset
df.head()
```

```python
df.shape
```

```python
# Get a summary of the dataset
print(df.info())
```

```python
# Display basic statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing values with N/A
df.fillna('N/A')
```

```python
df.columns
```

```python
df.dtypes
```

```python
# Convert the 'Age' column from float to int
df['age'] = df['age'].astype(int)
```

```python
# after convert float to int
df.dtypes
```

```python
# Display unique values for some columns
key_columns = ['sex','will_vote','education','income','job_type','company_type']
for column in key_columns:
    unique_values = df[column].unique()
    print(f"Unique values in the '{column}' column:")
    print(unique_values)
    print()  # Print a newline for better readability
```

```python
# Create a bar plot with 'Gender' on the x-axis and the count of 'will_vote' responses
plt.figure(figsize=(10, 6),facecolor='lightgray')
# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')

sns.countplot(data=df, x='sex', hue='will_vote')
plt.title('Count of Will Vote Responses by Gender',color='#2FC4A4',  fontproperties=font, fontsize= 18)
plt.xlabel('Gender', color='purple',fontproperties=font, fontsize= 18)
plt.ylabel('Count', color='violet', fontproperties=font, fontsize= 18)
plt.legend(title='Will Vote')
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6),facecolor='lightgray')  # Adjust size if necessary
# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')

plt.hist(df['type_of_city'], bins=3, color='red', edgecolor='black', alpha=0.7)
plt.xlabel('Type of City',color='#235176',fontproperties=font, fontsize= 18)
plt.ylabel('Frequency',color='#235176',fontproperties=font, fontsize= 18)
plt.title('Histogram of Type of City',color='#E2E931', fontproperties=font, fontsize= 18)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Rotate x-axis labels
plt.xticks(rotation=45)  # Adjust rotation angle as needed
plt.show()
```

```python
# List of columns to plot
key_columns = ['age']

# Plotting each column
for col in key_columns:
    plt.figure(figsize=(12, 10))
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
sns.lineplot(x = 'income', y = 'education', data = df, marker = 'o', color = 'purple')
```

```python
# Create a scatter plot to explore the relationship between 'job_type' and 'employment'
plt.figure(figsize=(10, 6),facecolor='#A6DDDA')
sns.scatterplot(x='job_type', y='employment', data=df)
# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here
plt.xlabel('job_type', color='red',  fontproperties=font, fontsize= 16)
plt.ylabel('employment', color='red',  fontproperties=font, fontsize= 16)
plt.title('Relationship between job_type and employment',  color='#3CEE1C',  fontproperties=font, fontsize= 18)
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
#Poll results by candidate and region
plt.figure(figsize=(12, 8))
sns.boxplot(x='candidate', y='weight1', hue='employment', data=df)
plt.title('Poll Results by Candidate and Region')
plt.xticks(rotation=45)
plt.show()
```

```python
num_col = ['age', 'weight1']
corr_df = df[num_col].corr()
corr_df
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(corr_df, annot=True, fmt=".2f", ax=ax, linewidths=0.5, linecolor="orange")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title("Correlation Between Different Variables")
plt.show()
```
